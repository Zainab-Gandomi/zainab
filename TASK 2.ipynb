```python
text = open('C:/Users/zainab/Desktop/shakespeare.txt')
```

```python
def split(data):
    return([i for item in data for i in item.split()])
```

```python
words = split(text)
```

```python
word_count = []
word_count= set(words)
len(word_count)
```

```python
DF['count'] = word_counts
```
